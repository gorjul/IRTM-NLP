```python
%matplotlib inline
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def print_top_words(model, feature_names, n_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic #%d:" % topic_idx)
        print(" ".join([feature_names[i]
                        for i in topic.argsort()[:-n_top_words - 1:-1]]))
    print()
```

```python
train_variants = pd.read_csv("test/training_variants")
test_variants = pd.read_csv("test/test_variants")
train_text = pd.read_csv("test/training_text", sep="\|\|", engine='python', header=None, skiprows=1, names=["ID","Text"])
test_text = pd.read_csv("test/test_text", sep="\|\|", engine='python', header=None, skiprows=1, names=["ID","Text"])
print('Unique Classes: ', len(train_variants.Class.unique()))
train_text.head()
```

```python
from sklearn.decomposition import LatentDirichletAllocation
lda = LatentDirichletAllocation(n_components=9, max_iter=5,
                                learning_method='online',
                                learning_offset=50.,
                                random_state=0)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
n_features = 50
tf_vectorizer = CountVectorizer(max_df=0.95, min_df=2,
                                max_features=50,
                                stop_words='english')
tf = tf_vectorizer.fit_transform(train_full['Text'])
```

```python
tf_feature_names = tf_vectorizer.get_feature_names()
tf_feature_names
```

```python
lda.fit(tf)
```

```python
print_top_words(lda, tf_feature_names, 10)
```
